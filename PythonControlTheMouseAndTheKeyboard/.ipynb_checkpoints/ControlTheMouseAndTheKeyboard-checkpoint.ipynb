```python
import pyautogui as py
```

增加保护措施，给pyautogui函数增加延迟为2.5s

```python
py.PAUSE = 2.5
```

鼠标输入

```python
# 获取屏幕大小
screenWidth, screenHeight = py.size()
print(screenWidth, screenHeight)
```

```python
# 获取鼠标位置
currentMouseX, currentMouseY = py.position()
print(currentMouseX, currentMouseY)
```

```python
# 将鼠标移动到 (1000, 500)
py.moveTo(1000, 500)
```

```python
# 鼠标点击左键两次，中间停留0.25s
py.click(clicks=1, interval=0.25)
```

```python
# 鼠标点击左键两次
py.doubleClick()
```

```python
# 鼠标向下移动200个像素
py.moveRel(None, 200)
```

键盘输入

```python
# 键盘输入，每个字母输入间隔为0.25s
py.typewrite('Hello world!', interval=0.25)
```

```python
# 按键esc等等
py.press('esc')
```

```python
# 按键左方向
py.press(['left', 'left'])
```

```python
# shift按下
py.keyDown('shift')
```

```python
# shift弹起
py.keyUp('shift')
```

```python
# 组合键
py.hotkey('ctrl', 'c')
```

鼠标滚轮

```python
py.scroll(-10)
```

```python
py.position()
```

屏幕截图

```python
# 屏幕截图
img = py.screenshot('my_screenshoot.png')
img
```

```python
# 移动到浏览器输入网址
py.moveTo(321, 73)
py.click()
py.typewrite("love.lrting.top", interval=0.25)
py.press('enter')
```
